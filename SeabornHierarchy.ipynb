```python
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.displot(df, x = 'total_bill'); #default distribution plot is histogram
```

```python
sns.displot(df, x= 'total_bill',kind='kde', fill = True, hue = 'day');
```

```python
sns.displot(df, x = 'total_bill', kind = 'ecdf');
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip'); # default relational plot is scatter plot
```

```python
df_by_dayofweek = df.groupby(['day'])['day','total_bill'].mean()

df_by_dayofweek = df_by_dayofweek.reset_index()

df_by_dayofweek.columns

df_by_dayofweek.columns = ['day','Average Bill Amount']

df_by_dayofweek

sns.relplot(data=df_by_dayofweek,x = 'day', y='Average Bill Amount',kind='line');
```

```python
sns.catplot(data=df, x = 'day', y = 'tip', kind = 'strip');
```

```python
sns.catplot(data=df, x = 'day', y = 'tip', kind = 'strip', hue = 'smoker');
```

```python
sns.catplot(data=df, x = 'smoker', y = 'total_bill', kind='swarm', hue = 'sex');
```

```python
sns.catplot(data=df, x = 'time', y = 'total_bill', kind = 'violin', hue = 'sex');
```

```python
sns.catplot(data=df, x = 'time', y='total_bill', kind = 'box');
```

```python
sns.catplot(data=df, x = 'time', y='total_bill', kind = 'point');
```

```python
sns.catplot(data = df, x = 'sex', y = 'total_bill', kind='bar');
```

```python
sns. relplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'day', hue='sex');
```
